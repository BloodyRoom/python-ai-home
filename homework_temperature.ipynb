```python
import numpy as np
```

```python
temperature = np.array([2, 5, 4, 6, 7, 3, 1]) 
```

```python
print("Середня температура: ", np.mean(temperature))
print("Максимальне значення: ", np.max(temperature))
print("Мінімальне значення: ", np.min(temperature))
print("Різниця між максимальним і мінімальним значенням: ", np.max(temperature) - np.min(temperature))
```
