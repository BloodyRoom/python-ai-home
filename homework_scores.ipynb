```python
import numpy as np
```

```python
math = np.array([80, 92, 75, 60, 88])
physics = np.array([70, 65, 90, 85, 78])
english = np.array([95, 100, 85, 90, 87])
```

```python
subjects = ["Math", "Physics", "English"]
```

```python
mathMean = np.mean(math)
mathMin = np.min(math)
mathMax = np.max(math)

physicsMean = np.mean(physics)
physicsMin = np.min(physics)
physicsMax = np.max(physics)

englishMean = np.mean(english)
englishMin = np.min(english)
englishMax = np.max(english)

subjects = np.array(["Math", "Physics", "English"])
averages = np.array([mathMean, mathMin, mathMax])
bestSubject = subjects[np.argmax(averages)]
```

```python
print("Math:")
print("\tСередній:", round(mathMean, 2))
print("\tМінімальний:", mathMin)
print("\tМаксимальний:", mathMax)

print("\nPhysics:")
print("\tСередній:", round(physicsMean, 2))
print("\tМінімальний:", physicsMin)
print("\tМаксимальний:", physicsMax)

print("\nEnglish:")
print("\tСередній:", round(englishMean, 2))
print("\tМінімальний:", englishMin)
print("\tМаксимальний:", englishMax)

print("\nПредмет з найвищим середнім:", bestSubject)
```
