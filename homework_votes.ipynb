```python
import numpy as np
```

```python
votes = np.array(["apple", "banana", "apple", "orange", "banana", "apple", "apple", "apple", "orange"])
```

```python
fruits, counts = np.unique(votes, return_counts=True)

leaderIndex = np.argmax(counts)
leader = fruits[leaderIndex]

sortedIndices = np.argsort(counts)[::-1]
top3Indices = sortedIndices[:3]
```

```python
print("Фрукт: Кількість")
for fruit, count in zip(fruits, counts):
    print(f"{fruit}: {count}")
```

```python
print("Лідер:", leader)
```

```python
print("Топ-3:")
for i in top3Indices:
    print(f"{fruits[i]}: {counts[i]}")
```
