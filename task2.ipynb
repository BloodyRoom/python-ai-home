```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ages = np.random.randint(18, 60, 100)

mean_age = np.mean(ages)

plt.figure()
plt.hist(ages)

plt.axvline(mean_age)

plt.xlabel('Вік')
plt.ylabel('Кількість людей')
plt.title('Розподіл віку')

plt.show()
```
