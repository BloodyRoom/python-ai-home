```python
import pandas as pd
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
data = pd.read_csv("reviews.csv")
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    return text
```

```python
data["cleaned_review"] = data["review"].apply(clean_text)
```

```python
stop_words = set(stopwords.words('english'))
```

```python
all_tokens_before = []
all_tokens_after = []
```

```python
processed_reviews = []

for review in data["cleaned_review"]:
    tokens = word_tokenize(review)

    all_tokens_before.extend(tokens)

    filtered_tokens = [
        token for token in tokens
        if token not in stop_words and len(token) >= 3
    ]

    all_tokens_after.extend(filtered_tokens)

    processed_reviews.append(filtered_tokens)
```

```python
data["tokens"] = processed_reviews

print("Оброблені токени:\n")
print(data[["tokens"]])

print("\nЗагальна кількість токенів ДО очищення:", len(all_tokens_before))
print("Загальна кількість токенів ПІСЛЯ очищення:", len(all_tokens_after))
```
