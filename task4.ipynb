```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
temperature = [20, 22, 19, 23, 25, 24, 21]
humidity = [60, 55, 65, 50, 45, 48, 52]

plt.figure()
plt.plot(days, temperature, label='Температура')
plt.plot(days, humidity, label='Вологість')

plt.xlabel('Дні')
plt.ylabel('Значення')
plt.title('Температура і вологість за тиждень')
plt.legend()

plt.show()
```
