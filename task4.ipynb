```python
import pandas as pd
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk import bigrams
from collections import Counter
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
data = pd.read_csv("reviews.csv")
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    return text

data["cleaned_review"] = data["review"].apply(clean_text)

stop_words = set(stopwords.words('english'))

all_tokens = []
```

```python
for review in data["cleaned_review"]:
    tokens = word_tokenize(review)

    filtered_tokens = [
        t for t in tokens
        if t not in stop_words and len(t) >= 3
    ]

    all_tokens.extend(filtered_tokens)
```

```python
bigram_list = list(bigrams(all_tokens))

bigram_freq = Counter(bigram_list)

top_10_bigrams = bigram_freq.most_common(10)

df_bigrams = pd.DataFrame(
    [(f"{w1} {w2}", count) for (w1, w2), count in top_10_bigrams],
    columns=["Bigram", "Frequency"]
)

print("Top 10 bigrams:\n")
print(df_bigrams)

df_bigrams.to_csv("feedback_bigrams.csv", index=False)

print("\nФайл 'feedback_bigrams.csv' збережено")
```
