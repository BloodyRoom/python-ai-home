```python
import numpy as np
```

```python
expenses = np.array([1200, 1500, 800, 600, 1000, 400, 900])
```

```python
sum = np.sum(expenses)
mean = np.mean(expenses)
max = np.max(expenses)
min = np.min(expenses)
daysAbove1000 = np.sum(expenses > 1000)
```

```python
print("Загальна сума витрат:", sum)
print("Середнє значення:", round(mean, 2))
print("Максимальна витрата:", max)
print("Мінімальна витрата:", min)
print("Кількість днів, коли витрати > 1000:", daysAbove1000)
```
