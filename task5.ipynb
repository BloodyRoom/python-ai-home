```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
hours = list(range(24))
load = np.random.randint(20, 100, 24)

plt.figure()
plt.plot(hours, load)

plt.fill_between(hours, load)

plt.grid()

plt.xlabel('Години')
plt.ylabel('Навантаження')
plt.title('Навантаження сервера за добу')

plt.show()
```
