```python
import matplotlib.pyplot as plt
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
plan = [100, 120, 140, 130, 150, 170]
fact = [90, 110, 135, 125, 160, 180]
```

```python
plt.figure()
plt.plot(months, plan, label='План')
plt.plot(months, fact, label='Факт')
plt.xlabel('Місяці')
plt.ylabel('Продажі')
plt.title('План, Факт продажів')
plt.legend()
plt.show()
```
