```python
import pandas as pd
import re
```

```python
reviews = [
    "The product is amazing and works perfectly!",
    "I really liked the fast delivery and good packaging.",
    "Terrible customer service, I am disappointed.",
    "The quality is excellent for this price.",
    "I will definitely buy from this store again.",
    "The item arrived damaged and late.",
    "Very friendly support team and quick response.",
    "The application is easy to use and helpful.",
    "Not worth the money, bad experience overall.",
    "Fantastic design and great performance.",
    "The battery life could be much better.",
    "I am satisfied with my purchase.",
    "Shipping was slow but the product is good.",
    "Excellent service and high quality materials.",
    "The website interface is clean and simple."
]
```

```python
df = pd.DataFrame({"review": reviews})
```

```python
df.to_csv("reviews.csv", index=False)

print("CSV файл 'reviews.csv' створено успішно.\n")
```

```python
data = pd.read_csv("reviews.csv")
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    return text

data["cleaned_review"] = data["review"].apply(clean_text)
```

```python
num_rows = len(data)
```

```python
data["word_count"] = data["cleaned_review"].apply(lambda x: len(x.split()))
average_length = data["word_count"].mean()
```

```python
print("Оброблені відгуки:\n")
print(data[["cleaned_review"]])

print("\nКількість рядків:", num_rows)
print("Середня довжина відгуків (у словах):", round(average_length, 2))
```
